```python
# 데이터프레임을 datetime으로 변환
df_gen['일시'] = pd.to_datetime(df_gen['일시'])

# 계절을 정의하는 함수
def get_season(date):
    year = date.year
    seasons = {
        '봄': (pd.Timestamp(year=year, month=3, day=1), pd.Timestamp(year=year, month=5, day=31)),
        '여름': (pd.Timestamp(year=year, month=6, day=1), pd.Timestamp(year=year, month=8, day=31)),
        '가을': (pd.Timestamp(year=year, month=9, day=1), pd.Timestamp(year=year, month=11, day=30)),
        '겨울': (pd.Timestamp(year=year, month=12, day=1), pd.Timestamp(year=year+1, month=2, day=28)),
    }
    for season, (start, end) in seasons.items():
        if start <= date <= end:
            return season
    return '겨울' if date.month == 12 else None

# 계절 열 추가
df_gen['계절'] = df_gen['일시'].apply(get_season)

# 지점별로 그룹화하고 계절별 MW/MWh의 평균 계산
both_result = df_gen.groupby(['지점', '계절'])['MW/MWh'].mean().unstack()
seasonal_result = df_gen.groupby('계절')['MW/MWh'].mean()

# 결과 출력
print(result)
print(seasonal_result)
```
